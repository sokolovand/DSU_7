# Задание 1

Изучить представленный набор данных на основе описания его столбцов, загрузить его и оставить 8 столбцов для дальнейшего изучения: surgery?, Age, rectal temperature, pulse, respiratory rate, temperature of extremities, pain, outcome.

```python
import pandas as pd
import numpy as np
```

```python
horse = pd.read_csv('horse_data.csv')
horse.head()
```

На основании просмотра описания стобцов установлено, кокой столбец за что отвечает

```python
horse_data = horse.iloc[:,[0,1,3,4,5,6,10,22]]
horse_data.rename(columns={
    '2': 'surgery',
    '1': 'age',
    '38.50': 'rectal_temperature',
    '66': 'pulse',
    '28': 'respiratory_rate',
    '3': 'temperature_of_extremities',
    '5': 'pain',
    '2.2': 'outcome'
}, inplace=True)

# Заменим '?' на np.NaN, а также приведем остальные значения в формат float
horse_data = horse_data.replace('?', np.NaN)
for col in horse_data.columns:
    
    horse_data[col] = horse_data[horse_data[col] != np.NaN][col].astype('float')

horse_data.head()
```

# Задание 2

Проанализировать значения по столбцам, рассчитать базовые статистики, найти выбросы.

#### Составим сводную таблицу со всеми параметрами и для каждого столбца

```python
# DF, с которым будет происходить объединение рассчитанных характеристик
df_stat_params = pd.DataFrame({
    'stat_method': ['nan(%)', 'min', 'max', 'mean', 'mode', 'median',
                   'var', 'std',  'q1', 'q3', 'iqr', 'lower_bound', 'upper_bound']
})
    
df_stat_params    
```

Сами характеристики. Циклом перебираются столбцы DF, игнорируются '?', вычисляются характеристики, образуется новый DF, по структуре идентичный df_stat_params, а дальше проиходит merge

```python
for col in horse_data.columns:
    
    template = horse_data[col]
    
    nan_ = round((template.isna().sum())/len(template)*100, 2)
    min_ = template.min()
    max_ = template.max()
    mean = round(template.mean(), 2)
    mode = template.mode()[0]
    median = template.median()
    var = round(template.var(), 2)
    std = round(template.std(), 2)
    q1 = round(template.quantile(0.25), 2)
    q3 = round(template.quantile(0.75), 2)
    iqr = q3 - q1
    lower_bound = round(q1 - 1.5 * iqr, 2)
    upper_bound = round(q3 + 1.5 * iqr, 2)
    
    params_for_1_col = [nan_, min_, max_, mean, mode, median, 
                           var, std, q1, q3, iqr, lower_bound, upper_bound]
    
#     # Для того, чтобы объединить 2 DF, используется столбец 'stat_method', который
#     # присутствует и в df_stat_params, и в new_df
    new_df = pd.DataFrame({'stat_method': ['nan(%)', 'min', 'max', 'mean', 'mode',
                                           'median', 'var', 'std',  'q1', 'q3', 'iqr', 
                                           'lower_bound', 'upper_bound'],
                           col: params_for_1_col})
    
    df_stat_params = df_stat_params.merge(new_df, how='left', on='stat_method')

df_stat_params
```

- У нас нет столбцов, в которых NaN элементов было бы больше 20%. Все они являются статистически значимыми
- Mode. Преимущественно взрослые особи. По шкале боли - 3 - преимущественно умеренные боли. outcome = 1 - чаще лошади выживали. Средний пульс выше нормы (норма 30-40)
- std. Довольно заметная изменчивость пульса. В 67% случаев пульс в диапазоне 71.93 +-28.68 ударов/мин

# Задание 3

Рассчитать количество пропусков для всех выбранных столбцов. Принять и обосновать решение о методе заполнения пропусков по каждому столбцу на основе рассчитанных статистик и возможной взаимосвязи значений в них. Сформировать датафрейм, в котором пропуски будут отсутствовать.

Количество пропусков в процентном соотношении представлено в таблице выше в строке nan(%)

Число пропусков:

```python
for col in horse_data.columns:
    try:
        nan = horse_data[col].value_counts()['?']
        print(f"-Число пропусков в столбце '{col}': {nan}")
        
    except:
        print(f"-В столбце '{col}' нет пропусков")
        continue
```

- Логически можно предположить, что столбцы "rectal_temperature",	"pulse", "respiratory_rate" имеют взаимосвязь. Если лошадь больна, то её температура поднимается, что может поднять пульс и частоту дыхания. Либо наоборот. Так как в столбцах respiratory_rate и rectal_temperature пропусков в 2 раза больше, чем в pulse, то попытаемся при заполнении NaN в respiratory_rate и rectal_temperature отталкиваться от значений в графе pulse. Для этого воспользуемся занятием по визуализации данных, чтобы посмотреть зависимости этих трех столбцов друг от друга


```python
import seaborn as sns

sns.pairplot(horse_data[['pulse', 'rectal_temperature', 'respiratory_rate']])
```

- Переменная rectal_temperature не сиьно-то и зависит от частоты пульса. Лишь когда превышает 100 ударов, виден подьем температуры. Частота дыхания respiratory_rate также не сильно зависима от пульса, этот параметр принимает разнообразные значения при разных наблюдениях. И сравнение температуры лошади с частотой дыхания также явной закономерности не показала. Данные варьируются значительно.

- Идея заполнить NaN через столбец pulse провалилась. Хотел, например, при пониженном пульсе установить значение, взятое с гафика. Скажем, пульс ниже - температура ниже, заполним NaN значение меньше среднего. Но так как явной связи нет, то пусть заместо пропусков стоит медиана

```python
horse_data['pulse'].fillna(horse_data['pulse'].median(), inplace=True)
horse_data['rectal_temperature'].fillna(horse_data['rectal_temperature'].median(), inplace=True)
horse_data['respiratory_rate'].fillna(horse_data['respiratory_rate'].median(), inplace=True)
```

- Для того, чтобы понять, как сильно поменялись данные из пункта 2, вызнана таблица со всеми статистиками

```python
round(horse_data.describe(), 2)
```

Несколько изменилось среднее значение столбцов pulse и respiratory_rate. Изменилось СКО данных, но тоже не очень сильно. Хоть это и зависит от целей поставленной задачи.

Тут уже труднее сделать выбор, так как категории 3 и 1 встречаются чаще всего, поэтому отдать 56 NaN значений одной категории будет вряд ли правильно. Нашел в интернете способ интерполяции категориальных данных, что и применил. В нем остался 1 NaN элемент, поэтому его просто заменим модой

```python
horse_data.temperature_of_extremities.fillna(horse_data.temperature_of_extremities.median(), inplace=True)
```

```python
horse_data.isna().sum()
```
