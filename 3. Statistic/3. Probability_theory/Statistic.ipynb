```python
import numpy as np
```

```python
reward_rules = {
    '777': 200,
    '999': 100,
    '555': 50,
    '333': 12,
    '111': 10,
    '77': 5,
    '7': 3,
    '00': 2,
    '0': 1
}
```

### Теоретический расчет

- Вероятность каждого события из тех, что позволют выиграть:                                           
999 - 1/1000                     
777 - 1/1000                         
555 - 1/1000                     
333 - 1/1000                     
111 - 1/1000                     
                                                               
- У "_77" старший разряд не может равняться 7, поэтому есть 9 вариантов                     
_77 - 9/1000                     
                                                               
- У "__7" второй разряд не может быть равен 7. 10 вариантов для первого разряда умножаем на 9 вариантов второго разряда                      
__7 - 9/100                     
                                                               
- Нули не ограничены свободными разрядами, поэтому для _00 есть 10 вариантов, а для __0 их уже 100                       
_00 - 1/100                     
__0 - 1/10                     

- Математическое ожидание

Теперь найдем произведение потенциального выигрыша на вероятность его появления.

На данный момент имеем, что:

Выигрыш 200 будет с вероятностью 1/1000                                     
Выигрыш 100 будет с вероятностью 1/1000                                     
Выигрыш 50 будет с вероятностью 1/1000                                     
Выигрыш 12 будет с вероятностью 1/1000                                     
Выигрыш 10 будет с вероятностью 1/1000                                     
Выигрыш 5 будет с вероятностью 9/1000                                     
Выигрыш 3 будет с вероятностью 9/100                                     
Выигрыш 2 будет с вероятностью 1/100                                     
Выигрыш 1 будет с вероятностью 1/10                                     

Теперь выигрыш умножаем на его вероятность, а затем складываем результат.

Результат равен: 0.807                                               
Это наше математическое ожидание, которое говорит, что на каждый потраченный рубль мы выигрываем примерно 0.8 рубля, что эквивалентно проигрышу 0.2 рубля при  каждой ставке.

### Моделирование

```python
trials = 10000000
reward = 0

for i in range(trials):

    number = str(np.random.randint(1000))

    for key in reward_rules:
        
        # Если последние len(key) символы совпали с ключом словаря наград, то прибавляется сумма. Благодаря break
        # дублирование наград отсутствует. После первого совпадения цикл прерывается.
        
        if number[-len(key):] == key:
            reward  += reward_rules[key]
            break
```

```python
# Отношение выигрыша к общей ставке
reward/trials
```

### Вывод

При 10e6 играх отношение выигрыша к проигрышу составляет почти 0.8, что очень похоже на рассчитанное математическое ожидание. По полученным результатам можно сказать, что данная игра на бесконечности будет убыточна для игрока. Временные победы, пусть и крупные, превратятся в убыток. В итоге удастся вернуть лишь 20% от своих ставок.
