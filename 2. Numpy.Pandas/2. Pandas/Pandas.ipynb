## Задание 1

Скачайте с сайта grouplens.org...movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
```

```python
movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')
ratings = ratings.merge(movies, on=['movieId'], how='left')
```

```python
ratings[(ratings['rating'] == 5.0)][['rating', 'title']].groupby('title').count().sort_values('rating', ascending=False).head()
```

## Задание 2

По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
power = pd.read_csv('power.csv')
```

```python
quantity = power[((power['country'] == 'Latvia') | (power['country'] == 'Litva') | (power['country'] == 'Estonia')) & (power['year'] >= 2005) & (power['year'] <= 2010) & ((power['category'] == 4) | (power['category'] == 12) | (power['category'] == 21)) & (power['quantity'] > 0)]['quantity'].sum()

# Что в этой длинной строке:
# 1. ((power['country'] == 'Latvia') | (power['country'] == 'Litva') | (power['country'] == 'Estonia')) Фильтр по 3м странам
# 2. & (power['year'] >= 2005) & (power['year'] <= 2010)     Фильтр года
# 3. & ((power['category'] == 4) | (power['category'] == 12) | (power['category'] == 21))   Фильтруем категории
# 4. & (power['quantity'] > 0)]['quantity'].sum()     Фильтруем измерения меньше нуля. Оставляем столбец "quantity"
# Суммируем

print(f'Суммарное потребление стран Прибалтики составляет {quantity} единиц энергии')
```
